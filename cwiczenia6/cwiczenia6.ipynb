```python
import pandas as pd
import requests
import json
```

```python
waluty = [
    'USD',
    'CHF',
    'EUR',
    'GBP',
    'JPY'
]
dane_walut = []

for waluta in waluty:
    url = f"http://api.nbp.pl/api/exchangerates/rates/A/{waluta}/last/25"
    response = requests.get(url, verify=True)
    if response.status_code != 200:
        raise Exception('Failed to load page {}'.format(url))

    data = response.json()
    rates = data['rates']
    
    for rate in rates:
        dane_walut.append({
            'code': waluta,
            'date': rate['effectiveDate'],
            'value': rate['mid']
        })

with open('dane_walut.json', 'w') as f:
    json.dump(dane_walut, f, ensure_ascii=False, indent=4)
```
